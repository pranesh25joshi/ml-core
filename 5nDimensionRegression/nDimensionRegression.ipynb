```python
import numpy as np
import pandas as pd
```

```python
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score
```

```python
from sklearn.datasets import make_regression
```

```python

x,y = make_regression(n_samples=100, n_features=2, n_targets=1, n_informative=2, noise=50 )
```

```python
df = pd.DataFrame({
    "feature1" : x[:,0],
    "feature2": x[:,1],
    "target": y
})
```

```python
import plotly.express as px
```

```python
fig = px.scatter_3d(df , x="feature1", y="feature2", z="target")
fig.show()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
print(mean_absolute_error(y_test, y_pred))
print(root_mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
lr.coef_
# this is the coefficient with the x1, x2 .... that are β1, β2 etc
```

```python
lr.intercept_
# this is the interecept b that is β0 for the place equation that cuts the 3D space with a place
```

```python
import plotly.graph_objects as go
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
xGrid, yGrid = np.meshgrid(y, x)

final = np.vstack((xGrid.ravel().reshape(1,100),yGrid.ravel().reshape(1,100))).T
z_final = lr.predict(final).reshape(10,10)

z = z_final
```

```python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target')

fig.add_trace(go.Surface(x = x, y = y, z =z ))

fig.show()
```
